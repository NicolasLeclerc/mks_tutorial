```python
from pymks.datasets import make_microstructure
import matplotlib.pyplot as plt
```

```python
data_blob = make_microstructure(n_samples=34, size=(201, 201), n_phases=2, grain_size=(30, 30), volume_fraction=[0.5, 0.5])
data_lamellar_v = make_microstructure(n_samples=33, size=(201, 201), n_phases=2, grain_size=(100, 10), volume_fraction=[0.5, 0.5])
data_lamellar_h = make_microstructure(n_samples=33, size=(201, 201), n_phases=2, grain_size=(10, 100), volume_fraction=[0.5, 0.5])
data = np.concatenate([data_blob, data_lamellar_v, data_lamellar_h], axis=0)
np.random.shuffle(data)
```

```python
fig = plt.figure(figsize=(12, 12))
for i in range(10):
    for j in range(10):
        ii = i * 10 + j
        ax = fig.add_subplot(10, 10, ii + 1)
        ax.set_aspect('equal')
        img = plt.imshow(data[ii])
        plt.axis('off')
        img.set_cmap('viridis')

plt.subplots_adjust(wspace=0.1, hspace=0.1)
plt.savefig('array.png')
```
